# **Blood Cell Clustering on Multiple Images**

### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
```

### Load and Process Multiple Images  

```python
image_folder = "blood_cell_images/"
```

```python
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]
```

### Display message based on image availability

```python
if image_files:
    print(f"Found {len(image_files)} images in '{image_folder}':")
    for img in image_files:
        print(f"   - {img}")  
else:
    print(f"No images found in '{image_folder}'. Please check the folder path")
```

### Define Image Processing Function  

```python
def process_image(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)

    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    features = []
    for contour in contours:
        M = cv2.moments(contour)
        if M["m00"] != 0:
            cX = int(M["m10"] / M["m00"])
            cY = int(M["m01"] / M["m00"])
            intensity = gray[cY, cX]
            features.append([cX, cY, intensity])

    return image_rgb, np.array(features)
```

### Apply DBSCAN Clustering  

```python
max_display = 50  
results = [] 

for idx, image_file in enumerate(image_files):
    image_path = os.path.join(image_folder, image_file)

    image_rgb, features = process_image(image_path)

    if len(features) > 0:
        dbscan = DBSCAN(eps=20, min_samples=5)
        labels = dbscan.fit_predict(features)
        num_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        num_noise = sum(labels == -1) 

        if idx < max_display:
            print(f"Processing: {image_file} | Clusters: {num_clusters} | Noise: {num_noise}")

            plt.figure(figsize=(10, 5))
            
            plt.subplot(1, 2, 1)
            plt.imshow(image_rgb)
            plt.title("Original Image")
            
            plt.subplot(1, 2, 2)
            for i, label in enumerate(set(labels)):
                color = "black" if label == -1 else plt.cm.jet(label / max(1, num_clusters))
                cluster_points = features[labels == label]
                plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=[color], label=f"Cluster {label}")
            
            plt.gca().invert_yaxis()
            plt.title(f"DBSCAN Clustering - {image_file}")
            plt.xlabel("X Coordinate")
            plt.ylabel("Y Coordinate")
            plt.legend()
            plt.show()
            
            results.append((image_file, num_clusters, num_noise, features, labels))  

    if idx >= max_display:
        break

print("Processing completed for 50 images.")
```

```python
for image_file, num_clusters, num_noise, features, labels in results:
    if num_clusters > 1:
        silhouette = silhouette_score(features, labels)  
        db_index = davies_bouldin_score(features, labels)
        ch_index = calinski_harabasz_score(features, labels)
    else:
        silhouette, db_index, ch_index = "N/A", "N/A", "N/A"

    print(f"\n📊 Metrics for {image_file}:")
    print(f"  - Clusters: {num_clusters}")
    print(f"  - Noise Points: {num_noise}")
    print(f"  - Silhouette Score: {silhouette}")
    print(f"  - Davies-Bouldin Index: {db_index}")
    print(f"  - Calinski-Harabasz Index: {ch_index}")
```
